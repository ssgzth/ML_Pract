```python
import pandas as pd
df=pd.read_csv("airquality.csv")
df.head(15)
```

```python
x=df['Ozone '].mean()
df['Ozone '].fillna(x,inplace=True)
df.head(15)
```

```python
for y in df.index:
  if df.loc[y,"Solar. R"]>250:
    df.loc[y,"Solar. R"]=150
df['Solar. R'].fillna(150,inplace=True)
df.head(15)
```

```python
for y in df.index:
  if df.loc[y,"Solar. R"]>190:
    z=df['Solar. R'].mean()
    df.loc[y,"Solar. R"]=z
df.head(15)
    
```
